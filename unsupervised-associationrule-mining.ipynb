**Remove return or replace item information**
 * remove invoice number with 'c'

```python
df_france = df_france[~df_france['Invoice'].str.contains('C')]
df_france.head()
```

```python
len(df_france['Invoice'].unique())
```

```python
len(df_france['Description'].unique())
```

```python
df_france.isnull().sum()
```

```python
df_france.groupby(['Invoice','Description'])['Quantity'].sum()
```

```python
basket = df_france.groupby(['Invoice','Description'])['Quantity'].sum().unstack().fillna(0)
basket.head()
```

```python
def trueorfalse(value):
    if value > 0:
        return True
    else:
        return False
```

```python
basket_tf = basket.applymap(trueorfalse)
basket_tf.head()
```

**Find most frequent items**

```python
from mlxtend.frequent_patterns import apriori,association_rules
frequent_itemsets = apriori(basket_tf, use_colnames =True , min_support = 0.08)
frequent_itemsets
```

**Apply Apriori Algorithm**

```python
rules1 = association_rules(frequent_itemsets , metric = 'lift' , min_threshold = 1)
rules1
```

```python
rules1.iloc[:,0:2].to_csv('rules1_invoice.csv')
```
